## Coreleation Matrix

```python
Data_sp500=pd.read_csv('sp500_px.csv')
```

```python
pd.set_option('display.max_columns',None)
Data_sp500.head()
```

```python
Data_CoMatrix=Data_sp500.loc[:,['T','CTL','FTR','VZ','LVLT']]
print(Data_CoMatrix.corr())
corrmat = Data_CoMatrix.corr(method='pearson')
f,ax = plt.subplots(figsize=(15, 15))
sns.heatmap(corrmat, ax=ax, cmap="coolwarm",annot=True)
```

```python
Data_sym=pd.read_csv('sp500_sym.csv')
```

```python
Data_sym.head()
```

```python
ETFs=Data_sp500.loc[(Data_sp500.Dates>'2012-07-01'),(Data_sym.symbol[(Data_sym.sector=='etf')&(Data_sym.symbol)])]
```

```python
corrmat = ETFs.corr(method='spearman')
f, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corrmat, ax=ax, cmap="YlGnBu", linewidths=0.1,annot=True)
```

```python
lc_loan=pd.read_csv('lc_loans.csv')
```

```python
lc_loan.head()
```

```python
lc_loan.dtypes
```

```python
x_tab=pd.crosstab(lc_loan.grade,lc_loan.status)
```

```python
total=x_tab.loc[:,'Charged Off':].sum(axis=1)
```

```python
x_tab=pd.concat([x_tab,total.rename('Total')],axis=1)
x_tab.head()
```

```python
x_tab.loc['Total']=pd.Series(x_tab.loc[:,'Charged Off':].sum(axis=0), index = x_tab.columns)
```
